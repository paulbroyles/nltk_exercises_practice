19\. ◑ What is the difference between the following two lines? Which one will give a larger value? Will this be the case for other texts?


    >>> sorted(set(w.lower() for w in text1))
    >>> sorted(w.lower() for w in set(text1))

The first line casts the words to lowercase before constructing the set, means that any given string of letters will only appear in the set once, regardless of its capitalization. The second line constructs a set of all the unique tokens (cap-sensitive) in text1 and then casts them to lowercase. Therfore the second will produce the larger value, because it will contain duplicates. The same will be true for other texts except in the very unlikely event that a text uses *no* words in another case and in lowercase.

```python
print(len(sorted(set(w.lower() for w in text1))))
print(len(sorted(w.lower() for w in set(text1))))
```

24\. ◑ Write expressions for finding all words in text6 that meet the conditions listed below. The result should be in the form of a list of words: ['word1', 'word2', ...].

<ol type="a">
<li>Ending in ize</li>
<li>Containing the letter z</li>
<li>Containing the sequence of letters pt</li>
<li>Having all lowercase letters except for an initial capital (i.e., titlecase)</li>
</ol>

```python
def find_words_with_letters(text, letters, where):
    if where == 'start':
        return [w for w in text if w.startswith(letters)]
    elif where == 'end':
        return [w for w in text if w.endswith(letters)]
    elif where == 'contains':
        return [w for w in text if letters in w]

def find_case(text, case):
    if case == 'upper':
        return [w for w in text if w.isupper()]
    elif case == 'lower':
        return [w for w in text if w.islower()]
    elif case == 'title':
        return [w for w in text if w.istitle()]
```

```python
find_words_with_letters(text6, 'ize', 'end')
```

```python
find_words_with_letters(text6, 'z', 'contains')
```

```python
find_words_with_letters(text6, 'pt', 'contains')
```

```python
find_case(text6, 'title')
```

27\. ◑ Define a function called vocab_size(text) that has a single parameter for the text, and which returns the vocabulary size of the text.

```python
def vocab_size(text):
    return len(set(w.lower() for w in text if w.isalpha()))

print(vocab_size(text1))
print(vocab_size(text2))
print(vocab_size(text3))
print(vocab_size(text4))
print(vocab_size(text5))
print(vocab_size(text6))
print(vocab_size(text7))
print(vocab_size(text8))
```

28\. ◑ Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.

```python
def percent(word, text):
    count = len([w for w in text if w.lower() == word.lower()])
    return 100 * count / len(text)

print(percent('love', text1))
print(percent('love', text2))
print(percent('love', text3))
print(percent('love', text4))
print(percent('love', text5))
print(percent('love', text6))
print(percent('love', text7))
print(percent('love', text8))
```
